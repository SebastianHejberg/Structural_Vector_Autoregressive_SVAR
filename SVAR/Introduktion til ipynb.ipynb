# Introduktion til Jupyter Notebook (med Markdown)

Jupyter Notebook er et interaktivt miljø, hvor du kan:
- Skrive og køre kode (f.eks. Python)
- Dokumentere dit arbejde med tekst i Markdown
- Vise visualiseringer og resultater direkte
- Skrive matematiske formler med LaTeX

### Overskrifter

```markdown
# H1
## H2
### H3
```

### Fed, kursiv, og mere
```markdown
**Fed tekst**
*Kursiv tekst*
~~Gennemstreget tekst~~
```

### Punktliste

```markdown
- Første punkt
- Andet punkt
  - Underpunkt
```

### Nummereret liste

```markdown
1. Først
2. Dernæst
3. Sidst
```

### Link

```markdown
[Besøg Jupyter](https://jupyter.org)
```

### Billede

```markdown
![Jupyter logo](https://jupyter.org/assets/main-logo.svg)
```

## To typer celler:

## Kør en celle (Markdown eller kode)


<div align="center">

| Handling                        | Windows / Linux           | Mac                     |
|----------------------------------|---------------------------|-------------------------|
| **Kør celle** og bliv i den     | `Ctrl + Enter`             | `Cmd + Return`           |
| **Kør celle** og gå til næste  | `Shift + Enter`            | `Shift + Return`         |
| **Kør celle og indsæt ny**     | `Alt + Enter`              | `Option + Return`        |

</div>


## `ESC`: "Command Mode"

Når du trykker `Esc`, går du i **Command Mode** – så kan du styre celler uden at skrive i dem. Tastaturgenveje fungerer kun i denne tilstand.

> Du kan trykke `Enter` igen for at vende tilbage til **Edit Mode** og begynde at skrive i cellen.
